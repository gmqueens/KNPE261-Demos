## Constant Error (CE): A Measure of Bias

Constant error (CE) provides a measure of the systematic bias in a performance outcome relative to a target. It indicates the average deviation of the outcomes from the target, allowing us to assess whether a performer tends to consistently overshoot or undershoot the target.

The formula for constant error is:

$$
CE = \frac{\sum (X_i - T)}{N}
$$

Where:

- $X_i$ = the outcome of each trial,
- $T$ = the target or goal,
- $N$ = the number of trials.


```python
## Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: outcomes of motor performance over multiple trials (X_i)
performance_outcomes = np.array([102, 98, 105, 99, 100, 101, 97, 104, 103, 99])

# Target value (T): the goal the performer is trying to achieve
target_value = 100

# Number of trials
num_trials = len(performance_outcomes)

# Plot performance data relative to target
plt.figure(figsize=(10, 6))
trials = np.arange(1, num_trials + 1)

# Plot the performance outcomes
plt.plot(trials, performance_outcomes, marker='o', label='Trials')

# Plot the target value as a horizontal line
plt.axhline(target_value, color='red', linestyle='--', label='Target')

# Add labels and title
plt.title('Constant Error Relative to Target', fontsize=14)
plt.xlabel('Trial', fontsize=12)
plt.ylabel('Score', fontsize=12)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Constant Error (CE): Measure of bias
constant_error = np.mean(performance_outcomes - target_value)
print(f"Constant Error (CE): {constant_error:.2f}")
```

## **Variable Error (VE): A Measure of Precision**

While constant error provides information about the average bias of performance, variable error measures the consistency or variability of performance outcomes across trials. Specifically, VE reflects how much the individual’s performance outcomes fluctuate around their own mean score, giving insight into the precision or stability of their actions.

VE is calculated as the standard deviation of the performance outcomes relative to the mean outcome:

$$
VE = \sqrt{\frac{1}{N} \sum (X_i - \bar{X})^2}
$$

Where:

- $X_i$ = the outcome of each trial,
- $\bar{X}$ = the mean of the performance outcomes,
- $N$ = the number of trials.

A high VE suggests that the performer is inconsistent, meaning that their performance fluctuates considerably from one trial to the next. Conversely, a low VE indicates that the performer consistently produces similar outcomes across trials, even if those outcomes are systematically biased (as indicated by CE).


```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: outcomes of motor performance over multiple trials (X_i)
performance_outcomes = np.array([102, 98, 105, 99, 100, 101, 97, 104, 103, 99])

# Number of trials
num_trials = len(performance_outcomes)

# Compute mean of the performance outcomes (X̄)
mean_outcome = np.mean(performance_outcomes)

# Plot performance data to show variability
plt.figure(figsize=(10, 6))
trials = np.arange(1, num_trials + 1)

# Plot performance outcomes as a line plot
plt.plot(trials, performance_outcomes, marker='o', label='Performance Outcomes')

# Plot mean outcome as a horizontal line
plt.axhline(mean_outcome, color='red', linestyle='--', label='Mean Outcome')

# Add labels and title
plt.title(' and Mean Outcome', fontsize=14)
plt.xlabel('Trial', fontsize=12)
plt.ylabel('Outcome', fontsize=12)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Calculate Variable Error (VE)
# VE is the standard deviation of the performance outcomes around their mean
variable_error = np.sqrt(np.sum((performance_outcomes - mean_outcome) ** 2) / num_trials)
print(f"Variable Error (VE): {variable_error:.2f}")
```

##**Total Variability (E): A Comprehensive Measure of Performance**

While constant error and variable error measure different aspects of performance, total variability (often denoted as **E** or total error) is a more holistic metric that accounts for both systematic bias and inconsistency. Total variability represents the overall spread of the performer’s results around the target and combines both the bias (as measured by **CE**) and the inconsistency (as measured by **VE**).

Mathematically, total variability is often calculated as the **root mean square error (RMSE)**, which represents the average magnitude of error, regardless of direction, between the performance outcomes and the target:

$$
E = \sqrt{\frac{1}{N} \sum (X_i - T)^2}
$$

Where:

- $ X_i $ = the outcome of each trial,
- $ T $ = the target or goal,
- $ N $ = the number of trials.

Total variability provides a global measure of the performer’s ability to achieve the target goal, accounting for both accuracy (CE) and precision (VE). It is useful when researchers or practitioners want to understand the overall error in performance without dis


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Sample data: outcomes of motor performance over multiple trials (X_i)
performance_outcomes = np.array([102, 98, 105, 99, 100, 101, 97, 104, 103, 99])

# Target value (T): the goal the performer is trying to achieve
target_value = 100

# Number of trials
num_trials = len(performance_outcomes)

# Plot performance outcomes relative to the target value
plt.figure(figsize=(10, 6))
trials = np.arange(1, num_trials + 1)

# Plot the performance outcomes
plt.plot(trials, performance_outcomes, marker='o', label='Performance Outcomes')

# Plot the target value as a horizontal line
plt.axhline(target_value, color='red', linestyle='--', label='Target Value')

# Add labels and title
plt.title('Performance Outcomes Relative to Target', fontsize=14)
plt.xlabel('Trial', fontsize=12)
plt.ylabel('Outcome', fontsize=12)

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

# Calculate Total Variability (E)
total_variability = np.sqrt(np.sum((performance_outcomes - target_value) ** 2) / num_trials)

# Output the Total Variability (E)
print(f"Total Variability (E): {total_variability:.2f}")
```

```python
import requests
from bs4 import BeautifulSoup # To be able to obtain data from a table
import pandas as pd

# Fetch the webpage
url = 'https://www.basketball-reference.com/players/j/jamesle01/gamelog/2004#pgl_basic'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Locate the specific table
table = soup.find('table', {'id': 'pgl_basic'})

# Convert the HTML table into a DataFrame
df = pd.read_html(str(table))[0]

# Remove any duplicate headers that might be present in the middle of the table
df = df[df['Rk'] != 'Rk']

# Drop rows that are completely empty
df = df.dropna(how='all')

# Display the first few rows to verify
print(df.head())

# Convert the 'PTS' column to numeric, coercing errors in case of non-numeric values
df['PTS'] = pd.to_numeric(df['PTS'][0:10], errors='coerce')

# Calculate the average points per game
average_ppg = df['PTS'].mean()

# Calculate the standard deviation of points per game
std_dev_ppg = df['PTS'].std()

# Display the results
print(f"Average Points Per Game: {average_ppg:.2f}")
print(f"Standard Deviation of Points Per Game: {std_dev_ppg:.2f}")
```
